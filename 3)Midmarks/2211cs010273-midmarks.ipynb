# Dataset Report:
**Midterm Marks Dataset Report**

**1. Overview**
The dataset consists of **718 records** and **8 columns**, capturing student performance in midterm exams for various subjects. The data is taken from the "SEM2 MID 1 - ALPHA" sheet.

**2. Data Fields**
- **S.NO** – Serial number (601 non-null values, some missing entries).
- **SECTION** – Class section (691 non-null values, mostly "ALPHA").
- **DV** – Marks for "DV" subject (716 non-null values).
- **M-II** – Marks for "M-II" subject (716 non-null values).
- **PP** – Marks for "PP" subject (716 non-null values).
- **BEEE** – Marks for "BEEE" subject (716 non-null values).
- **FL** – Marks for "FL" subject (715 non-null values).
- **FIMS** – Marks for "FIMS" subject (716 non-null values).

**3. Observations**
- The dataset has **some missing values**, especially in the "S.NO" and "SECTION" columns.
- The marks appear to be stored as **text values**, requiring conversion to numerical format for proper analysis.
- The dataset primarily focuses on **midterm marks** for different subjects.

**4. Recommendations**
- **Data Cleaning:** Convert all marks to numeric format and handle missing values.
- **Statistical Analysis:** Compute mean, median, and distribution of marks.
- **Visualization:** Create histograms and box plots to understand performance trends.
- **Section-wise Analysis:** Evaluate performance across different sections if more sections exist.

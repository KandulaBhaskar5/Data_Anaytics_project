### Dataset Description:
The dataset consists of **86 records** and **12 columns**, representing students' marks on various questions in an assessment.

#### **Columns:**
1. **Total** – The total marks obtained by the student.
2. **Q1aM4** – Marks for Question 1a (out of 4).
3. **Q1bM6** – Marks for Question 1b (out of 6).
4. **Q2aM6** – Marks for Question 2a (out of 6).
5. **Q2bM4** – Marks for Question 2b (out of 4).
6. **Q3aM5** – Marks for Question 3a (out of 5).
7. **Q3bM5** – Marks for Question 3b (out of 5).
8. **Q4aM3** – Marks for Question 4a (out of 3).
9. **Q4bM7** – Marks for Question 4b (out of 7).
10. **Q5M10** – Marks for Question 5 (out of 10).
11. **Q6aM4** – Marks for Question 6a (out of 4).
12. **Q6bM6** – Marks for Question 6b (out of 6).

#### **Observations:**
- The **"Total"** column contains integer values, while the rest contain floating-point values.
- Some columns have missing values, indicating that not all students attempted every question.
- The dataset primarily focuses on individual question scores and the total marks.

```python
import matplotlib.pyplot as plt
import pandas as pd

# Load dataset
df = pd.read_csv("class_marks.csv")

# Summing question parts while handling missing values
df["Q1"] = df[["Q1aM4", "Q1bM6"]].sum(axis=1, skipna=True)
df["Q2"] = df[["Q2aM6", "Q2bM4"]].sum(axis=1, skipna=True)
df["Q3"] = df[["Q3aM5", "Q3bM5"]].sum(axis=1, skipna=True)
df["Q4"] = df[["Q4aM3", "Q4bM7"]].sum(axis=1, skipna=True)
df["Q5"] = df["Q5M10"]  # Already a single column
df["Q6"] = df[["Q6aM4", "Q6bM6"]].sum(axis=1, skipna=True)

# Drop old columns
df.drop(["Q1aM4", "Q1bM6", "Q2aM6", "Q2bM4", "Q3aM5", "Q3bM5", 
         "Q4aM3", "Q4bM7", "Q5M10", "Q6aM4", "Q6bM6"], axis=1, inplace=True)

# Calculate total marks per question
question_totals = df[["Q1", "Q2", "Q3", "Q4", "Q5", "Q6"]].sum()

# Plot pie chart
plt.figure(figsize=(8, 8))
plt.pie(question_totals, labels=question_totals.index, autopct="%1.1f%%", colors=["#ff9999","#66b3ff","#99ff99","#ffcc99","#c2c2f0","#ffb3e6"])
plt.title("Percentage Contribution of Each Question to Total Marks")
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

# Load dataset
df = pd.read_csv("class_marks.csv")

# Summing question parts while handling missing values
df["Q1"] = df[["Q1aM4", "Q1bM6"]].sum(axis=1, skipna=True)
df["Q2"] = df[["Q2aM6", "Q2bM4"]].sum(axis=1, skipna=True)
df["Q3"] = df[["Q3aM5", "Q3bM5"]].sum(axis=1, skipna=True)
df["Q4"] = df[["Q4aM3", "Q4bM7"]].sum(axis=1, skipna=True)
df["Q5"] = df["Q5M10"]  # Already a single column
df["Q6"] = df[["Q6aM4", "Q6bM6"]].sum(axis=1, skipna=True)

# Calculate total score
df["Total"] = df[["Q1", "Q2", "Q3", "Q4", "Q5", "Q6"]].sum(axis=1)

# Assign grades based on total score (Modify as per your grading system)
def assign_grade(score):
    if score >= 90:
        return "A"
    elif score >= 75:
        return "B"
    elif score >= 60:
        return "C"
    elif score >= 45:
        return "D"
    else:
        return "F"

df["Grade"] = df["Total"].apply(assign_grade)

# Melt the dataframe for seaborn's boxplot (long format)
df_melted = df.melt(id_vars=["Grade"], value_vars=["Q1", "Q2", "Q3", "Q4", "Q5", "Q6"], 
                     var_name="Question", value_name="Score")

# Box plot visualization by Grade
plt.figure(figsize=(12, 6))
sns.boxplot(x="Question", y="Score", hue="Grade", data=df_melted, palette="Set2")

# Formatting
plt.title("Box Plot of Scores for Each Question by Grade")
plt.xlabel("Questions")
plt.ylabel("Scores")
plt.legend(title="Grade")
plt.grid(True)

# Show plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

# Load the dataset
df = pd.read_csv("class_marks.csv")

# Calculate missing values per column
missing_counts = df.isnull().sum()
missing_counts = missing_counts[missing_counts > 0]  # Only keep columns with missing values

# Check if there are missing values
if missing_counts.empty:
    print("No missing values in the dataset.")
else:
    # Pie chart visualization
    plt.figure(figsize=(8, 8))
    plt.pie(missing_counts, labels=missing_counts.index, autopct="%1.1f%%", 
            colors=sns.color_palette("coolwarm", len(missing_counts)), startangle=140)
    plt.title("Proportion of Missing Values per Column")
    plt.legend(title="Columns", loc="best")
    plt.show()
```
